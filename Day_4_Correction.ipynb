 Pandas Part:
	3.	Create a DataFrame with all columns.
	4.	Convert Diabetes into a categorical column with labels: 'No' (0) and 'Yes' (1)
	5.	Categorize Age into:
		'Young': < 35

	7.	Sort the data by BloodPressure in descending order and show top 5 patients.

	9.	Save the cleaned DataFrame to a CSV file (e.g., "cleaned_patients.csv").

```python
import pandas as pd
import numpy as np
import random
```

```python
Age = np.linspace(18, 90, 30) # Not random
random.shuffle(Age)
```

```python
BP = np.random.normal(120,  15, size = 30)
```

```python
Gender = np.random.choice(["Male", "Female"], 30)
```

```python
Cholesteral = np.linspace(150, 300, 30)
random.shuffle(Cholesteral)
```

```python
Diabetes = (BP> 120) & (Cholesteral>200).astype(int)
```
